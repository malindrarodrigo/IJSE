```python
import pandas as pd
df = pd.read_csv('music.csv')
df.shape
```

```python
df.values
```

### dividing input and output

```python
X=df.drop(columns=['genre']) # outputset
X
```

```python
y= df['genre'] #output data
y
```

#### we are using DecisionTreeClassifire to clasify our data

```python
pip install -U scikit-learn
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
model.fit(X,y)
predictions = model.predict([[21,1],[28,0]])
predictions 
```

### teating the model

```python
from sklearn.model_selection import train_test_split # 
X_train, X_test , y_train , y_test = train_test_split(X,y, test_size =0.2)
```

### creating model with train data

```python
model.fit(X_train,y_train)
predictions = model.predict(X_test)
```

### teating the predictions with testdata

```python
from sklearn.metrics import accuracy_score
```

```python
score = accuracy_score (y_test,predictions)
score
```

### Persisting Models

```python
import joblib
model = DecisionTreeClassifier()
model.fit(X.values,y.values)
joblib.dump(model, 'music-recommender.joblib')
```

```python
model =joblib.load ('music-recommender.joblib')
predictions = model.predict([[21,1],[22,0]])
predictions
```
